#### Scatterplot

```python
wdir = 'X:/user/liprandicn/mt-comparison/'
temp_type = "heat"
unit = "total"
age_group = "all"
years = range(2010,2021)
cause = "All causes"
burkart = utils.LoadScatter(wdir, "/burkart2022/output/ComparisonBurkart/MOR_ComparisonBurkart_SSP2_ERA5_ISO3_1980-2022", years, temp_type, unit, age_group, cause)
carleton =  utils.LoadScatter(wdir, "carleton2022\\output\\ComparisonCarleton/ISO3/MOR_ComparisonCarleton_SSP2_ERA5_NoAdap_1980-2022", years, temp_type, unit, age_group, cause)
zhao = pd.read_excel(f'{wdir}/zhao2021/output/zhao2021_tab1.xlsx', header=[0,1], index_col=0).xs("heat-related", level=0, axis=1).sort_index()
# ballester = utils.LoadBallesterScatter(wdir)

models = [zhao, burkart, ballester, carleton]#, scovronick]
models_name = ["Zhao et al., 2021", "Burkart et al., 2021", "Romanello et al., 2025", "Carleton et al., 2022"]#, "Scovronick et al., 2024"]
color_pallete = ["#5B9279", "#022B3A", "#E3B448", "#6F1A07"]#, "#707A5E"]

fig, axs = plt.subplots(1,2,figsize=(7, 5), gridspec_kw={'width_ratios': [5,1]})
axs = axs.flatten()

for df, name, color in zip(models, models_name, color_pallete):
    axs[0].errorbar(x=df.index[:-1], 
                y=df['mean'][:-1], 
                yerr=[df['p5'][:-1],df['p95'][:-1]], 
                fmt='o',         
                capsize=5,       
                color=color, 
                ecolor=color,   
                markersize=8,
                label=name)
    
    axs[1].errorbar(x=df.index[-1:], 
                y=df['mean'][-1:], 
                yerr=[df['p5'][-1:],df['p95'][-1:]], 
                fmt='o',         
                capsize=5,       
                color=color, 
                ecolor=color,   
                markersize=10,
                label=name)

for i in range(2):
    axs[i].set_facecolor('whitesmoke')
    axs[i].grid(True,  color="white")
    axs[i].yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x/1000:,.0f}'))
    for spine in axs[i].spines.values():
        spine.set_visible(False)
    
# axs[0].set_ylim(0,6e5)
axs[0].set_ylabel("Mortality (thousand people)", fontsize=12)
axs[0].legend()
axs[1].yaxis.tick_right()
axs[1].yaxis.set_label_position("right")

plt.suptitle('Annual average excess deaths due to heat for all ages', fontsize=14)
plt.tight_layout()

plt.savefig(wdir+"figures/IAMC/scatter_comparison.pdf", bbox_inches="tight")
plt.show()
```
