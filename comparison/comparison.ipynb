```python
import utils
import numpy as np
import pandas as pd
import xarray as xr
import geopandas as gpd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

### All ages comparisson

#### Scatterplot

```python
wdir = 'X:/user/liprandicn/mt-comparison/'
temp_type = "heat"
unit = "total"
age_group = "all"
years = range(2010,2021)
cause = "All causes"
burkart = utils.LoadScatter(wdir, "/burkart2022/output/ERA5_no-counter/mortality_ERA5_no-counter_SSP2_ERA5_ISO3_2000-2025", years, temp_type, unit, age_group, cause)
carleton =  utils.LoadScatter(wdir, "carleton2022\\output\\default\\mortality_default_SSP2_ERA5_ISO3_NoAdap_2000-2025", years, temp_type, unit, age_group, cause)
zhao = (
    pd.read_excel(f'{wdir}/zhao2021/output/zhao2021_tab1.xlsx', header=[0,1], index_col=0)
    .xs("heat-related", level=0, axis=1)
)

models = [zhao, burkart, ]
models_name = ["Zhao et al., 2021", "Burkart et al., 2022", "Scovronick et al., 2024"]
color_pallete = ["#022B3A", "#A8763E", "#6F1A07"]

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('whitesmoke')

for df, name, color in zip(models, models_name, color_pallete):
    plt.errorbar(x=df.index, 
                y=df['mean'], 
                yerr=[df['p5'],df['p95']], 
                fmt='o',         
                capsize=5,       
                color=color, 
                ecolor=color,   
                markersize=8,
                label=name)

plt.title('Average mortality (2010-2020)')
plt.ylabel("Total estimated mortality")
plt.grid(True,  color="white")
plt.legend()
plt.ylim(0,6e5)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
def format_100k(x, pos):
    return f'{x/1000:g}k'
ax.yaxis.set_major_formatter(FuncFormatter(format_100k))

plt.show()
```
